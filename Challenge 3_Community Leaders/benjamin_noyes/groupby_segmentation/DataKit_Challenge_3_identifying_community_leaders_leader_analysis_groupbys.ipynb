```python
wefarm = pd.read_parquet("/content/drive/MyDrive/DataKit_WeFarm/wefarm.parquet")
```

```python
wefarm["response_sent_dt"] = pd.to_datetime(wefarm["response_sent"], format="ISO8601")
wefarm["question_set_dt"] = pd.to_datetime(wefarm["question_sent"], format="ISO8601")
```

```python
def parse_datetime_col(df, colstr):
  df[f'{colstr}_year'] = df[colstr].dt.year
  df[f'{colstr}_month'] = df[colstr].dt.month
  df[f'{colstr}_day'] = df[colstr].dt.day
  df[f'{colstr}_hour'] = df[colstr].dt.hour
  df[f'{colstr}_minute'] = df[colstr].dt.minute
  df[f'{colstr}_second'] = df[colstr].dt.second
  df[f'{colstr}_day_name'] = df[colstr].dt.day_name()
  df[f'{colstr}_day_of_week'] = df[colstr].dt.day_of_week
  df[f'{colstr}_month_name'] = df[colstr].dt.month_name()
  df[f'{colstr}_quarter'] = df[colstr].dt.quarter
  df[f'{colstr}_week'] = df[colstr].dt.isocalendar().week
  df[f'{colstr}_date'] = df[colstr].dt.date
  df[f'{colstr}_time'] = df[colstr].dt.time

  return df
```

```python
wefarm = parse_datetime_col(wefarm, "response_sent_dt")
wefarm = parse_datetime_col(wefarm, "question_set_dt")
```

# speed

```python
wefarm["question_response_min_diff"] = (wefarm["question_set_dt"] - wefarm["response_sent_dt"]).dt.total_seconds() / 60
```

```python
wefarm["question_response_time_diff"] = wefarm["question_set_dt"] - wefarm["response_sent_dt"]
```

```python
wefarm["response_sent_dt_buck"] = pd.qcut(wefarm["question_response_min_diff"], 100).cat.codes
```

```python
rtime_user_id_gb = wefarm.groupby("response_user_id"
  ).agg({"response_sent_dt_buck": "mean"}
  ).sort_values(by="response_sent_dt_buck", ascending=True)
```

```python
rt_3_user_id_lst = rtime_user_id_gb[rtime_user_id_gb["response_sent_dt_buck"] < 5].index
rt_2_user_id_lst = rtime_user_id_gb[
  (rtime_user_id_gb["response_sent_dt_buck"] > 5) &
  (rtime_user_id_gb["response_sent_dt_buck"] <= 15)
].index
rt_1_user_id_lst = rtime_user_id_gb[
  (rtime_user_id_gb["response_sent_dt_buck"] > 15) &
  (rtime_user_id_gb["response_sent_dt_buck"] <= 40)
].index
rt_0_user_id_lst = rtime_user_id_gb[(rtime_user_id_gb["response_sent_dt_buck"] > 40)].index
```

```python
rt_segment_dct = {
  "0": rt_0_user_id_lst.tolist(),
  "1": rt_1_user_id_lst.tolist(),
  "2": rt_2_user_id_lst.tolist(),
  "3": rt_3_user_id_lst.tolist()
}
```

# conversation starters

```python
question_response_count_gb = wefarm.groupby("question_id").count().sort_values(
  by="question_user_id", ascending=False
)
question_id_response_boundary_lst = question_response_count_gb[question_response_count_gb["question_user_id"]<=20].index
```

```python
wefarm_sub = wefarm[wefarm["question_id"].isin(question_id_response_boundary_lst)]
```

```python
wefarm_sorted = wefarm_sub.sort_values(["question_id", "question_response_min_diff"])
```

```python
wefarm_sorted["response_order"] = wefarm_sorted.groupby("question_id").cumcount()+1
```

```python
wefarm_sorted["total_responses"] = wefarm_sorted.groupby("question_id")["response_id"].transform("count")
```

```python
wefarm_sorted["responses_after"] = wefarm_sorted["total_responses"] - wefarm_sorted["response_order"]
```

```python
wefarm_sorted.loc[:20, ["question_id", "question_response_min_diff", "response_order", "total_responses", "responses_after"]]
```

```python
cs_user_id_gb = wefarm_sorted.groupby("response_user_id"
  ).agg({"responses_after": "mean"}
  ).sort_values(by="responses_after", ascending=False)
```

```python
cs_0_user_id_lst = cs_user_id_gb[cs_user_id_gb["responses_after"] <= 1.0].index
cs_1_user_id_lst = cs_user_id_gb[
  (cs_user_id_gb["responses_after"] > 1.0) &
  (cs_user_id_gb["responses_after"] <= 3.0)
].index
cs_2_user_id_lst = cs_user_id_gb[
  (cs_user_id_gb["responses_after"] > 3.0) &
  (cs_user_id_gb["responses_after"] <= 8.0)
].index
cs_3_user_id_lst = cs_user_id_gb[(cs_user_id_gb["responses_after"] > 8.0)].index
```

```python
cs_segments_dct = {
  "0":cs_0_user_id_lst.tolist(),
  "1":cs_1_user_id_lst.tolist(),
  "2":cs_2_user_id_lst.tolist(),
  "3":cs_3_user_id_lst.tolist()
}
```

```python
[len(j) for j in cs_segments_dct.values()]
```

# tenure

```python
ten_user_id_gb0 = wefarm.groupby("question_user_id").agg({
  "question_set_dt": ["min", "max"]
})

ten_user_id_gb1 = wefarm.groupby("response_user_id").agg({
  "response_sent_dt": ["min", "max"]
})
```

```python
ten_user_id_gb = pd.concat([ten_user_id_gb0, ten_user_id_gb1], axis=1)
ten_user_id_gb.columns = ['_'.join(col) for col in ten_user_id_gb.columns]
```

```python
ten_user_id_gb["earliest"] = ten_user_id_gb[[
    "question_set_dt_min",
    "question_set_dt_max",
    "response_sent_dt_min",
    "response_sent_dt_max"
]].min(axis=1)
```

```python
ten_user_id_gb["latest"] = ten_user_id_gb[[
    "question_set_dt_min",
    "question_set_dt_max",
    "response_sent_dt_min",
    "response_sent_dt_max"
]].max(axis=1)
```

```python
ten_user_id_gb["tenure"] = (ten_user_id_gb["latest"] - ten_user_id_gb["earliest"]).dt.days
```

```python
ten_0_user_id_lst = ten_user_id_gb[ten_user_id_gb["tenure"] <= 1].index
ten_1_user_id_lst = ten_user_id_gb[
  (ten_user_id_gb["tenure"] > 1) &
  (ten_user_id_gb["tenure"] <= 20)
].index
ten_2_user_id_lst = ten_user_id_gb[
  (ten_user_id_gb["tenure"] > 20) &
  (ten_user_id_gb["tenure"] <= 400)
].index
ten_3_user_id_lst = ten_user_id_gb[(ten_user_id_gb["tenure"] > 400)].index
```

```python
ten_segments_dct = {
  "0": ten_0_user_id_lst.tolist(),
  "1": ten_1_user_id_lst.tolist(),
  "2": ten_2_user_id_lst.tolist(),
  "3": ten_3_user_id_lst.tolist()
}
```

```python
[len(j) for j in ten_segments_dct.values()]
```
